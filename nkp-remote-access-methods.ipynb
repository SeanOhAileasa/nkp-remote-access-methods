######  1. Virtual Private Network (VPN)
***

After completing this topic, you should be able to: ``use a VPN including IPsec, SSL/TLS/DTLS, site-to-site, and client-to-site in a scenario``

In this presentation, we'll take a closer look at VPNs, or the Virtual Private Network. 

And this is used to provide secure connectivity across an unsecure network domain using encryption. 

And, of course, the unsecure network to which we are referring is the Internet. It is the most public network there is, and as such, it is not secure.

But the common use cases include accessing a business network remotely. So when you have traveling users or people who work from home, they might need access to the resources of the internal organization. And as such, they need a virtual private network connection. 

And, in fact, it might even go the other way. Maybe the person who works from home is actually in the central site, we'll say. They might then need to access their system at home, so you can go the other direction if necessary. Now that's up to you in terms of how you set this up. 

But, again, the idea is that in terms of connectivity, there is a network that exists between wherever you are when you're traveling and/or wherever you live and your business environment. And that network is the Internet, so let's take advantage of it, let's use it. But, of course, we need to do so in a private manner because it's a public network.

######  IPsec - Secured Implementation of the IP Protocol
***

[IPsec] So this is where the encryption comes into play. 

Now IPsec, or IP sec, is simply a secured implementation of the IP protocol. And as such, it's ideal for use on the Internet. 

But as mentioned, we need to use the Internet in a private fashion, so we need to encrypt the communications that we are sending back and forth. 

So it's a set of protocols used to secure the communication. And it protects the data that is exchanged between an IPsec gateway and any remote network or any host that supports the IPsec protocol. 

So we see a remote user that is essentially the IPsec client. It negotiates a connection with an IPsec VPN device or a firewall and they, in essence, establish that communication and it creates what's known as a tunnel. And anyone that is able to intercept the traffic, basically only gets encrypted information. So they basically cannot gain access to what the information is. So that's how you use this public network in a private fashion. 

Now, IPsec also allows for isolation with respect to the internal resources. You can create these permitted subnets, so that, essentially, you can say once you have connected to the VPN device or the firewall, you may now access the resources of this permitted subnet. And that allows you to isolate those resources so that you can access them, but perhaps not access other systems that might be a little more secure. 

So that's up to you, however you configure that, but it is supported. So, again, this just handles the encryption so that you can communicate securely.

######  Site-to-site VPN
***

######  2. Remote Desktop Protocol (RDP)
***

After completing this topic, you should be able to: ``use RDP in a scenario``

Now in this next section, we'll take a look at several protocols used to establish remote connectivity to other systems. And we'll begin with RDP, or the Remote Desktop Protocol.

And this is actually a proprietary protocol developed by Microsoft. 

And it provides users with the ability to access a remote computer's desktop over any network connection, including the Internet. Now that requires a little additional configuration, but it can still be done. 

And, again, the idea is that you don't just establish a connection to access a shared file or a shared printer or something along those lines, you quite literally see the desktop of the remote computer. So this is something that many network administrators live by. They need to administer many servers that could be quite geographically separated. So as long as you have network connectivity to those servers, you can simply access the remote desktop and perform, virtually, any task. At least from a software perspective, there is nothing you can't do. You might as well be sitting at that computer. 

Now this is actually something that evolved from an earlier Microsoft service known as Terminal Services. And even that, essentially, evolved from some of the oldest methods of connecting to remote systems way back in the days of mainframes. 

Effectively, the mainframe did everything. It did all the processing, all the storage was accessed from the mainframe. But you had terminals that were connected to the mainframe so that you could at least interact with them. So you might have a keyboard and a screen in front of you, but you are remotely connecting to the mainframe to give it the instructions and tell it what to do.

So, in other words, the terminal really did nothing. It might have looked like what we tend to think of as a computer. But they were actually referred to as dumb terminals because they didn't do anything but provide an interface. 

So Terminal Services was essentially the same thing, connecting to a remote computer. All the processing, all the work, is being done on the remote system. 

You aren't really doing anything except accessing an interface, and that, again, carries through to the Remote Desktop Protocol. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Remote Desktop Client Utility
***

You have a small piece of software on your client, the Remote Desktop client utility. 

That is simply used to establish and maintain the connection. But once you are connected to that remote desktop, everything you do happens on the remote system. 

So your system becomes little more than a terminal. 

But, of course, you can run other applications along with the RDP connection. So your system, of course, is no longer a dumb terminal. 

But once you make the connection to that remote system, whatever it is that you do happens on the remote system. 

So this provides secure network communications, as well. And, as mentioned, it's ideal for remote management. 

So that you can remotely diagnose and resolve problems, whether they be with servers or maybe it's for the benefit of telecommuters who are working from home or maybe from a branch office. But, again, the idea is to simply be able to access that system in its entirety to do whatever it is you need to do on that remote system without having to actually go anywhere.

###### Features
***

It offers several features, such as encryption, to ensure that the data being transferred is secure. 

It supports smart card authentication, if you are using that. 

It supports compression for bandwidth reduction, resource sharing so that the remote system can actually make use of the resources that you have on your own local system. For example, if I connect to the remote system and open a document and I want to print that document, I can print it on my local printer that is attached to my own local system. So that's resource sharing. 

Multiple displays are supported. And you can even disconnect without logging off. So that if you are going to reconnect, let's say, in a short amount of time, it can maintain your session if desired.

###### Protocol Distinction - Remote Desktop Connection (RDC) versus Remote Assistance
***

Now, one other distinction with respect to the protocol is, really, just how it's used. 

Officially, there are two applications in Windows environments. 

Remote Desktop Connection

The Remote Desktop Connection, which, again, just allows you to log on to a Windows computer remotely and access its entire desktop.

But then there is also Remote Assistance, and this incorporates components of RDP, if you will, but its purpose is a little different. It allows you to connect to an existing Windows session to effectively provide, as the name indicates, assistance, technical support. 

So, typically, the way that you distinguish the two is that when you use a Remote Desktop Connection, it's usually into a server, it doesn't have to be. But it's usually an administrator remotely accessing a server to perform administrative tasks. And as such, there usually is no one else using that server. There's nobody sitting at the console doing anything. And, in fact, that is the case with most servers. You generally don't use them in a day-to-day fashion. 

But Remote Assistance is someone who is actually sitting at the computer and needs help. So you can establish a Remote Assistance connection. You can see the desktop and you can even take control if necessary, but it's basically to walk a person through a problem. If they don't know how to access a certain feature or perform a certain task, you can tell them, launch this application, click here, enter that. In essence, instruct them or walk them through the process so that they can complete the task. 

But, again, the idea is that they are there performing the task, they are still logged in as themselves. You can just see that as it's happening. And as mentioned, if necessary you can take control, but you don't really log in to the system. That person is logged in as themselves. 

And when you establish a Remote Desktop Connection, you log in as yourself. So that's another difference as well, the identity of the user. 

Now, both of these features can be enabled or disabled at the discretion of administration, so you do not have to have them on. 

Servers tend to have them enabled by default. And, usually, the Professional Editions or the Business Editions of the desktops in Windows support it, but it might not be enabled. 

But one of the differences is that servers support multiple connections by default, clients only support one. 

But Remote Assistance is usually disabled by default on all systems so you do have to enable it. 

Now you can do both through something like a group policy, wherein you can distribute the settings with a policy that basically says, these systems all have it off or those systems all have it on, for either feature. But that, again, is up to you. 

But, ultimately, Remote Desktop Connectivity is a very necessary feature in most environments these days to simply be able to establish connections to remote systems from any location.

######  3. Secure Shell (SSH) - ``Port 22``
***

After completing this topic, you should be able to: ``use SSH in a scenario``

In this presentation, we'll take a look at our next remote connectivity protocol, SSH, or Secure Shell. 

And very much like the Remote Desktop Protocol that we just saw, this provides a secure way to access a computer system remotely. 

It establishes authenticated and secure encrypted data communications. 

And it's commonly used by administrators for managing servers and the applications on those servers remotely. 

It allows you to log in to those systems. You can then perform maintenance, troubleshoot various issues, and of course, execute commands. 

So, again, very similar to the Remote Desktop Protocol. But this uses a standard client/server model, wherein you have a client application that is used to establish a connection into an SSH server. Now, regardless of the official classification of client versus server, the client is simply whichever system you used to initiate the session and the server is any system that accepts the incoming requests. 

So you do not necessarily have to be designated as a server type of operating system. It's any system that is accepting the connection. 

Now that is essentially the same in the Remote Desktop Protocol, as well, because even if you have two desktop systems, say, running Windows 10, you can connect from one to the other. So the client is whichever one you used to initiate the connection, and the server is whichever one accepts that request. So, again, client/server in terms of the classification of the operating system is a little bit skewed. 

But as for SSH itself, it uses port 22 by default to communicate, but you can change that if you want to. And it is available in most operating systems by default, but primarily Unix and Linux. And this is where you find the most significant difference between SSH and something like Remote Desktop Protocol. 

Recall that RDP is a Microsoft proprietary protocol, hence, that's the reason you don't normally see it installed on Windows. Now it is available for Windows, but it's not there by default, RDP is. So you can certainly download SSH-based applications for Windows, if you are familiar with an application known as PuTTY, that is a very common SSH-based utility that can be installed on Windows.

######  5. Network Terminal Protocol (Telnet)
***

After completing this topic, you should be able to: ``use telnet in a scenario``

Our next remote connectivity protocol is Telnet, and this is a network terminal emulation protocol. 

The name doesn't really stand for anything, it really is just a name. But the TE does tend to indicate that terminal emulation. 

But this is a user command for accessing computers remotely. So again, administrators can access someone else's computer remotely to perform troubleshooting or various tasks. 

And the commands that you enter into the Telnet program are certainly entered on your own computer. But they are executed on the remote computer as if you're on the remote system's console. 

So again, this hearkens back to those early days that I mentioned in a previous presentation about mainframes. Everything is actually executed on the mainframe. The terminal was simply the means by which you could interface with it. So no matter what you typed in on your screen, nothing happened on your system. It was simply sent back to the mainframe and executed there. So that's the idea of terminal emulation. I am entering the command on my own system but it gets sent to the remote system and executes there. So it allows the remote system to be controlled and to even communicate with other systems on a network if desired. 

So again, this does sound a awful lot like RDP. But one of the core differences is that Telnet is entirely command driven. There is no graphic interface at all with Telnet. It's a very old protocol, it has been around for a long time. So no graphic component whatsoever with Telnet.

######  Public Key Infrastructure - One-way Communication
***

[PKI â€“ One way] So with respect to that implementation, PKI can be implemented in either one-way communication or two-way.

But when it's a one-way communication, it is still this asymmetric system where it uses two keys to encrypt the overall communication. The public key is used to encrypt the data. And the private key is the only thing that can be used to then decrypt the information that was encrypted with the public. 

Now, the keys themselves are based on exceptionally large prime numbers, and I mean exceptionally large, hundreds of digits. 

And the reason for this is that if you take two exceptionally large prime numbers, you get an exceptionally, exceptionally large semi-prime as an answer. And it just simply means that the answer was created by these two prime numbers, that's a semi-prime. 

And semi-primes have the unique characteristic that the only way to figure out which two original numbers were used to create this number is to try every possible number. 

There's no algorithm that you can run them through that says, this number x and this number y were the two factors that were used. You have to try every single possible number to get at the answer, thatâ€™s the essence of PKI. The only way to crack the decryption is to, in short, try every possible combination. 

And the numbers of combinations are so high that itâ€™s just practically impossible. It would take thousands of years for a very high powered computer to run through every possible combination, so it's just not practical. 

So anyway, with the one-way traffic, the sender, before sending a plain text message, requests the public key of the recipient. So if I'm the sender and you are the recipient, I request your public key. Anyone can get a public key, but that is the means to encrypt the data. The private key is held only by you, and it essentially knows what it sent you as the public key. 

But again, anything that is encrypted with the public can only be decrypted with the private. And since you are the only one who holds the private, you are the only entity that is able to decrypt it. So you encrypt it with the public, you send it on its way. If someone were to intercept that, the only thing they can get is a public. The public key cannot be used to decrypt data that was encrypted with a public key. You then use your private key to access the data and reverse the encryption. And you are presented with the original plain text. But that's only useful in a one-way transaction.

######  Public Key Infrastructure - Two-way Communication
***

[PKI â€“ Two way] If it's two-way, and we both need to talk to each other, then we, in fact, need a third key.

But I don't protect the original message with the public key, I generate a third key known as a symmetric key. 

This is another encryption algorithm. And I protect that algorithm with the public key. 

If we are going to communicate two ways in an encrypted manner, we both need to use the same key. We both need to speak the same language, if you will. So I generate a new key, I protect it with your public, and I send it off to you. But again, from the perspective of an attacker, if they were to intercept it, they can get a public key. But a public key is useless in decrypting anything that was encrypted with a public key. So the symmetric key remains protected. 

So in terms of Internet banking, the bank would be the recipient, you would be the sender. The bank is the only entity that maintains the private key. Therefore, they are the only ones able to access the symmetric key that you just sent them, and that secures the two-way communications.
